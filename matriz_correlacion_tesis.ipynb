```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
from google.colab import drive
drive.mount('/content/drive')
```

```python
ruta="/content/drive/MyDrive/tesis/dataset.XLSX"
data=pd.read_excel(ruta)
```

```python
data.dropna(inplace=True)
```

```python
df_num=data.loc[:,data.columns!="pit"].astype("float64")
dataset=pd.concat([df_num,data["pit"]],axis=1,join="outer")
dataset.info()
```

dataset.corr() devuelve un dataframe con informacion sobre los coeficientes de correlacion de -1, 0 y 1.
-1 Y 1 existe una relación fuerte de las variables. El signo nos indica la dirección. -1 indica una pendiente negativa y 1 indica una pendiente positiva.
Entre mas se acerce al valor absoluto de 1, más fuerte será la relación. Entre más se acerce al valor absoluto de 0 es una correlación baja.

```python
corr=dataset.corr()
fig,ax=plt.subplots(figsize=(20,20))
sns.heatmap(data=corr,cmap="RdYlGn",annot=True,linewidths=0.01,linecolor="black",square=True,ax=ax)
```
